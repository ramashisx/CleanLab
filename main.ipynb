```python
import cv2 as cv
import pandas as pd
```

```python
import os
import shutil
```

```python
base = "train"
for file in os.listdir(base):
    img = cv.imread(f"{base}/{file}")
    histg = cv.calcHist([img],[0],None,[256],[0,256])
    if histg[255] < 400:
        shutil.copy(f"{base}/{file}", f"clean/{file}")
```

```python
df = pd.read_csv("train-labels.csv")
```
